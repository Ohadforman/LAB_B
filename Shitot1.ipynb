## $ z = e^{i\theta}\\ $  ,  $ w(z) = \sqrt{z} $   

## $ U = cos(\frac{\theta}{2}) $ , $V = sin(\frac{\theta}{2})$ , $\theta\in[0,2\pi]$

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
# Data for plotting
theta = np.linspace(0,2*np.pi,301)
x = np.cos(theta/2)
y = np.sin(theta/2)
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set(xlabel='$U$', ylabel='$V$', title=r'$U+Vi$')
ax.grid()
#fig.savefig("test.png")
plt.show()
```

## z = $2+e^{i\theta}$

## $ U = \Re(\sqrt(2+e^{i\theta}))\\ $  ,  $ V = \Im(\sqrt(2+e^{i\theta}))\\ $ , $\theta\in[0,2\pi]$

```python
x = np.real((np.cos(theta)+2+np.sin(theta)*np.complex(0,1))**0.5)
y = np.imag((np.cos(theta)+2+np.sin(theta)*np.complex(0,1))**0.5)
fig, ax = plt.subplots()
axes = plt.gca()
axes.set_xlim([min(x),max(x)])
axes.set_ylim([min(y),max(y)])
ax.plot(x, y)
ax.set(xlabel='$U$', ylabel='$V$', title=r'$U+Vi$')
ax.grid()
#fig.savefig("test.png")
plt.show()
```
