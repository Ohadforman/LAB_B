## $ u = cos(\frac{\theta}{2}) $ , $v = sin(\frac{\theta}{2})$ , $\theta\in[0,2\pi]$

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
# Data for plotting
theta = np.linspace(0,2*np.pi,301)
x = np.cos(theta/2)
y = np.sin(theta/2)
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set(xlabel='U', ylabel='V', title='U+Vi')
ax.grid()
#fig.savefig("test.png")
plt.show()
```
